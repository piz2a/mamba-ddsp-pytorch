```python
import torch
import torch.nn as nn
```

```python
x = torch.randn(2, 3, requires_grad=True)
x
```

```python
y = torch.fft.fft(x)
y
```

```python
# grad can be implicitly created only for scalar outputs
# therefore, y.backward() is impossible.
loss = (y.abs()**2).sum()
loss
```

```python
loss.backward()
x.grad
```
